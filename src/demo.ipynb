```python
stable_tfs = measure_camera_movement("../data/stable.avi", verbose=False, every_n=2, stop_frame=np.infty)
```

```python
gleicher2_tfs = measure_camera_movement("../data/gleicher2.mp4", verbose=False, every_n=2, stop_frame=np.infty)
```

```python
stable_dx = []
stable_dy = []
stable_timeline = []
for tf in stable_tfs:
    stable_dx.append(tf.dx)
    stable_dy.append(tf.dy)
    stable_timeline.append(tf.timeline)

gleicher2_dx = []
gleicher2_dy = []
gleicher2_timeline = []
for tf in gleicher2_tfs:
    gleicher2_dx.append(tf.dx)
    gleicher2_dy.append(tf.dy)
    gleicher2_timeline.append(tf.timeline)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
# stable video
ax1.plot(stable_timeline, stable_dx, label="dx")
ax1.plot(stable_timeline, stable_dy, label="dy")
ax1.set_title("Stable")
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Pixels")
ax1.legend()
# gleicher2
ax2.plot(gleicher2_timeline, gleicher2_dx, label="dx")
ax2.plot(gleicher2_timeline, gleicher2_dy, label="dy")
ax2.set_title("Gleicher2")
ax2.set_xlabel("Time (s)")
fig.set_size_inches(8, 3.5)
plt.tight_layout()
```
